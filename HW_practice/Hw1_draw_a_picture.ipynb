```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

plt.title("A Heart Of Love")
N = np.linspace(-2, 2, 1000);
Y1 = np.sqrt(2 * np.sqrt(np.square(N))-np.square(N));
Y2 = (-2.14) * np.sqrt(np.sqrt(2) - np.sqrt(abs(N)));
plt.plot(N,Y1,color = "red")
plt.plot(N,Y2,color = "red")
plt.show()
```
