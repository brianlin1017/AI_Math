```python
from ipywidgets import interact
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000)

def myplot(n=1):
    plt.plot(x, np.sin(n*x))
```

```python
interact(myplot, n=(1, 50));
```

```python
%matplotlib inline
x = range(10)
y1 = np.linspace(0, 10, 10)
y2 = np.linspace(0, 5, 10)
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
plt.figure(figsize = (5, 5))
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
plt.figure(figsize = (5, 5), facecolor = "b") 
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
plt.figure(figsize = (5, 5))
plt.plot(x, y1, label = "y1")
plt.plot(x, y2, label = "y2")

plt.legend(loc = "upper left")
```

```python
plt.figure(figsize = (5, 5))
plt.plot(x, y1, label = "y1")
plt.plot(x, y2, label = "y2")

plt.legend(loc = "best")
```

```python
#import numpy as np
plt.figure(figsize = (5, 5))
plt.plot(x, y1, label = "y1")
plt.plot(x, y2, label = "y2")

plt.legend(loc = "best")

plt.xlabel("xxxx")
plt.ylabel("yyyy")
plt.title("title")

plt.xlim(-2, 20)
plt.xlim(-20, 10)

plt.xticks(np.arange(-2, 20, 5))
plt.yticks(np.arange(-20, 10, 5))
```
